Given the input and output variables of the problem, we can consider 5 __linguistic variables:__  
1. __(D)ifficulty__  
2. __(I)mportance__  
3. __(R)emaining work__  
4. __(T)ime to exam__  
5. __(P)riority__  

First 4 rules are the `antecedents` of the fuzzy system and the last one is the `consequent`.    
Let's start with a more concrete description of every LV.  

````python
lv_difficulty = LinguisticVariable(name="Difficulty", ling_values_dict={
        "easy": LinPWMF([0, 1], [4, 0]),
        "medium": TrapMF(0, 4, 7, 10),
        "hard": LinPWMF([7, 0], [10, 1])
    })
LinguisticVariableViewer(lv_difficulty).show()
````

````python
from fuzzy_systems.core.linguistic_variables.two_points_lv import TwoPointsPDLV

lv_importance = TwoPointsPDLV("Importance", 0, 10)
LinguisticVariableViewer(lv_importance).show()
````

````python
lv_remaining_work = LinguisticVariable("RemainingWork", ling_values_dict={
    "few": LinPWMF([0, 1], [25, 0]),
    "medium": TrapMF(0, 25, 75, 100),
    "much": LinPWMF([75, 0], [100, 1])
})

LinguisticVariableViewer(lv_remaining_work).show()
````

````python
lv_time = LinguisticVariable("TimeToExam", ling_values_dict={
    "no_time": LinPWMF([0, 1], [30, 0]),
    "still_have_time": LinPWMF([0, 0], [30, 1])
})

LinguisticVariableViewer(lv_time).show()
````

````python
from fuzzy_systems.core.linguistic_variables.three_points_lv import ThreePointsLV
lv_priority = ThreePointsLV("Priority", 0, 0.5, 1)
LinguisticVariableViewer(lv_priority).show()
````

The choice of the shapes of the membership functions is based on the intuition. For example, we decided that the lingsuistic variable "Difficulty" should have the "medium" linguistic value with a considerable range of values, this is why we chose the trapezoidal membership function for it.  
Another factor that influenced our choices was that we wanted to try all possible shapes for the LVs to see how they influence the resulting predicted values exactly.  

__Fuzzy Rules:__  
`rule1:` IF "Difficulty" is "easy" AND "RemainingWork" is "few" THEN "Priority" is "low"  
`rule2:` IF "RemainingWork" is "medium" OR "TimeToExam" is "still_have_time" THEN "Priority" is "medium"  
`rule3:` IF "Importance" is "high" AND "RemainingWork" is NOT "few" THEN "Priority" is "high"  

To test our system we first create only 3 fuzzy rules that cover all possible linguistic values of the output ("Priority") 

````python
# Implementation of the fuzzy rules described above 
r1 = FuzzyRule(
    ants=[
        Antecedent(lv_difficulty, "easy"),
        Antecedent(lv_remaining_work, "few")
    ],
    ant_act_func=AND_min,
    cons=[
        Consequent(lv_priority, "low"),
    ],
    impl_func=MIN
)

r2 = FuzzyRule(
    ants=[
        Antecedent(lv_remaining_work, "medium"),
        Antecedent(lv_time, "still_have_time")
    ], 
    ant_act_func=OR_max,
    cons=[
        Consequent(lv_priority, "medium")
    ], 
    impl_func=MIN
)

r3 = FuzzyRule(
    ants=[
        Antecedent(lv_importance, "high"),
        Antecedent(lv_remaining_work, "few", is_not=True)
    ],
    ant_act_func=AND_min,
    cons=[
        Consequent(lv_priority, "high"),
    ],
    impl_func=MIN
)

# Create the FIS
fis = FIS(
    rules=[r1, r2, r3],
    aggr_func=np.max, 
    defuzz_func=COA_func 
)

# Execute the FIS 
input_values = {'Difficulty': 8, 'Importance': 4, "RemainingWork": 30, "TimeToExam": 4}
predicted_value = fis.predict(input_values)

# View the FIS
fisv = FISViewer(fis, figsize=(12, 10))
fisv.show()
````

````python
# Execute all tests
input_values_list = [
    {'Difficulty': 8, 'Importance': 4, "RemainingWork": 30, "TimeToExam": 4},
    {'Difficulty': 4, 'Importance': 6, "RemainingWork": 50, "TimeToExam": 6},
    {'Difficulty': 10, 'Importance': 7, "RemainingWork": 10, "TimeToExam": 10},
    {'Difficulty': 2, 'Importance': 3, "RemainingWork": 80, "TimeToExam": 3},
    {'Difficulty': 6, 'Importance': 9, "RemainingWork": 40, "TimeToExam": 14},
    {'Difficulty': 8, 'Importance': 8, "RemainingWork": 70, "TimeToExam": 3},
    {'Difficulty': 3, 'Importance': 3, "RemainingWork": 30, "TimeToExam": 15},
    {'Difficulty': 10, 'Importance': 10, "RemainingWork": 100, "TimeToExam": 1},
    {'Difficulty': 1, 'Importance': 1, "RemainingWork": 0, "TimeToExam": 30}
]

for in_val in input_values_list:
    predicted_value = fis.predict(in_val)
    print(in_val, "=>", predicted_value)
````

As we can see, some of the results seem to be reasonable. But most of the predicted values are concentrated around the priority of 0.5. To make the FIS more sensible, lets add some more fuzzy rules and rebuild the FIS.  

`rule4:` IF "Difficulty" is "easy" AND "Importance" is "low" AND "RemainingWork" is "few" AND "TimeToExam" is "sill_have_time" THEN "Priority" is "low"  
`rule5:` IF "Difficulty" is "hard" AND "Importance" is "high" AND "RemainingWork" is "much" AND "TimeToExam" is "no_time" THEN "Priority" is "high"  

These two rules are the most "obvious" we can have. Their goal is to enforce the extreme situations (for example, situations where we immediatly need to start preparing for the exam or when there is no need to bother at all)

````python
r4 = FuzzyRule(
    ants=[
        Antecedent(lv_difficulty, "easy"),
        Antecedent(lv_importance, "low"),
        Antecedent(lv_remaining_work, "few"),
        Antecedent(lv_time, "still_have_time")
    ],
    ant_act_func=AND_min,
    cons=[
        Consequent(lv_priority, "low"),
    ],
    impl_func=MIN
)

r5 = FuzzyRule(
    ants=[
        Antecedent(lv_difficulty, "hard"),
        Antecedent(lv_importance, "high"),
        Antecedent(lv_remaining_work, "much"),
        Antecedent(lv_time, "no_time")
    ],
    ant_act_func=AND_min,
    cons=[
        Consequent(lv_priority, "high"),
    ],
    impl_func=MIN
)

fis = FIS(
    rules=[r1, r2, r3, r4, r5],
    aggr_func=np.max, 
    defuzz_func=COA_func 
)

# Execute the FIS 
input_values = {'Difficulty': 8, 'Importance': 4, "RemainingWork": 30, "TimeToExam": 4}
predicted_value = fis.predict(input_values)

fisv = FISViewer(fis, figsize=(12, 10))
fisv.show()

for in_val in input_values_list:
    predicted_value = fis.predict(in_val)
    print(in_val, "=>", predicted_value)
````

After adding the two rules described above, we can see that the result predictions do not change at all. The reason is that the `rule2` has the most importance for the resulting membership function (Priority aggregated), since the aggregation function is max and the activation function between antecedents is OR_max.  
This is why we will try to modify the `rule2` not to let it dominate the others. We will bring it more specific. 

`rule2:` IF ("RemainingWork" is "few" OR "TimeToExam" is "still_have_time") AND "Importance" is "high" THEN "Priority" is "medium"  

Since we can only use one fuzzy operator per rule, we will split `rule2` into multiple fuzzy rules.  

`rule2a:` IF "RemainingWork" is "few" AND "Importance" is "high" THEN "Priority" is "medium"  
`rule2b:` IF "TimeToExam" is "still_have_time" AND "Importance" is "high" THEN "Priority" is "medium"  



````python
r2a = FuzzyRule(
    ants=[
        Antecedent(lv_remaining_work, "few"),
        Antecedent(lv_importance, "high"),
    ],
    ant_act_func=AND_min,
    cons=[
        Consequent(lv_priority, "medium"),
    ],
    impl_func=MIN
)

r2b = FuzzyRule(
    ants=[
        Antecedent(lv_time, "still_have_time"),
        Antecedent(lv_importance, "high"),
    ],
    ant_act_func=AND_min,
    cons=[
        Consequent(lv_priority, "medium"),
    ],
    impl_func=MIN
)

fis = FIS(
    rules=[r1, r2a, r2b, r3],
    aggr_func=np.max, 
    defuzz_func=COA_func 
)

input_values = {'Difficulty': 10, 'Importance': 10, "RemainingWork": 100, "TimeToExam": 1}
predicted_value = fis.predict(input_values)

fisv = FISViewer(fis, figsize=(12, 10))
fisv.show()

for in_val in input_values_list:
    predicted_value = fis.predict(in_val)
    print(in_val, "=>", predicted_value)
````

The results are already better, but still the predicted values for the extreme cases are not satisfying, because they are not extreme enough. To enforce the extreme cases, we can change the "low" and "high" values of the "Priority" linguistic variable  

````python
lv_priority = LinguisticVariable("Priority", ling_values_dict={
    "low": LinPWMF([0, 1], [0.25, 0]),
    "medium": TrapMF(0, 0.25, 0.75, 1),
    "high": LinPWMF([0.75, 0], [1, 1])
})
LinguisticVariableViewer(lv_priority).show()

r1 = FuzzyRule(
    ants=[
        Antecedent(lv_difficulty, "easy"),
        Antecedent(lv_remaining_work, "few")
    ],
    ant_act_func=AND_min,
    cons=[
        Consequent(lv_priority, "low"),
    ],
    impl_func=MIN
)

r3 = FuzzyRule(
    ants=[
        Antecedent(lv_importance, "high"),
        Antecedent(lv_remaining_work, "few", is_not=True)
    ],
    ant_act_func=AND_min,
    cons=[
        Consequent(lv_priority, "high"),
    ],
    impl_func=MIN
)


r2a = FuzzyRule(
    ants=[
        Antecedent(lv_remaining_work, "few"),
        Antecedent(lv_importance, "high"),
    ],
    ant_act_func=AND_min,
    cons=[
        Consequent(lv_priority, "medium"),
    ],
    impl_func=MIN
)

r2b = FuzzyRule(
    ants=[
        Antecedent(lv_time, "still_have_time"),
        Antecedent(lv_importance, "high"),
    ],
    ant_act_func=AND_min,
    cons=[
        Consequent(lv_priority, "medium"),
    ],
    impl_func=MIN
)

fis = FIS(
    rules=[r1, r2a, r2b, r3],
    aggr_func=np.max, 
    defuzz_func=COA_func 
)

input_values = {'Difficulty': 10, 'Importance': 10, "RemainingWork": 100, "TimeToExam": 1}
predicted_value = fis.predict(input_values)

fisv = FISViewer(fis, figsize=(12, 10))
fisv.show()

for in_val in input_values_list:
    predicted_value = fis.predict(in_val)
    print(in_val, "=>", predicted_value)
````

As we can see, the extreme high value of the priority increased of about 4%, but the modificatons we have made influenced some other predicted results. For instance, we have around 28/100% of priority where we should have the extreme low (around 0) value.  
To conclude, we are satisfied of our results. They match the intuitive expectation we have.
